```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
DATASET_PATH = './Normalized_Data.csv'
```

```python
data = pd.read_csv(DATASET_PATH)
data.head()
```

```python
classes = sorted(list(data.Label.unique()))
labels = [classes.index(l) for l in data['Label']]
print("got labels")
data = data.drop(['Label'],axis=1)
print("dropped label")
data = data.fillna(data.mean())
print("filled nans")
print(classes)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.3, random_state=42)
x_train.head()
```

```python
arr = x_train.to_numpy()
np.where(np.isnan(arr))
```

```python
rf = RandomForestRegressor(n_estimators = 10, random_state = 42)
rf.fit(x_train, y_train)
```

```python
predictions = rf.predict(x_test)
```

```python
print("Accuracy")
print(len([i for i in range(len(predictions)) if predictions[i]==y_test[i]])/len(predictions))
print("Accuracy Per class")
print("Class\tAccuracy")
for i in range(len(classes)):
    class_i = [x for x in range(len(y_test)) if y_test[x]==i]
    correct = [x for x in range(len(class_i)) if predictions[x]==y_test[x]]
    print(classes[i]+"\t"+str(len(correct)/len(class_i)))
```
